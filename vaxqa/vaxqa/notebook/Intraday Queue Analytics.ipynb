```python

df2 = pd.DataFrame(columns=[
    'NoPerDay',
    'RegistrationTime',
    'VaccineTime',
    'WaitingTime',
    'RegistrationDesks',
    'VaccineDesks',
    'SeatingCap',
    'W1',
    'W2',
    'W3',
    'QueueOutside',
    'measure',
])

df_avg = df[[
    'NoPerDay',
    'RegistrationTime',
    'VaccineTime',
    'Waitingtime',
    'RegistrationDesks',
    'VaccineDesks',
    'SeatingCap',
    'Avg_W1',
    'Avg_W2',
    'Avg_W3',
    'Avg_QueueOutside',
]].rename(columns={'Waitingtime': 'WaitingTime', 'Avg_W1': 'W1', 'Avg_W2': 'W2', 'Avg_W3': 'W3', 'Avg_QueueOutside': 'QueueOutside', })

df_avg['measure'] = 'Avg'
df2 = df2.append(df_avg, ignore_index=True)

df_90 = df[[
    'NoPerDay',
    'RegistrationTime',
    'VaccineTime',
    'Waitingtime',
    'RegistrationDesks',
    'VaccineDesks',
    'SeatingCap',
    '90_W1',
    '90_W2',
    '90_W3',
    '90_QueueOutside',
]].rename(columns={'Waitingtime': 'WaitingTime', '90_W1': 'W1', '90_W2': 'W2', '90_W3': 'W3', '90_QueueOutside': 'QueueOutside', })

df_90['measure'] = '90'
df2 = df2.append(df_90, ignore_index=True)

df_95 = df[[
    'NoPerDay',
    'RegistrationTime',
    'VaccineTime',
    'Waitingtime',
    'RegistrationDesks',
    'VaccineDesks',
    'SeatingCap',
    '95_W1',
    '95_W2',
    '95_W3',
    '95_QueueOutside',
]].rename(columns={'Waitingtime': 'WaitingTime', '95_W1': 'W1', '95_W2': 'W2', '95_W3': 'W3', '95_QueueOutside': 'QueueOutside', })

df_95['measure'] = '95'
df2 = df2.append(df_95, ignore_index=True)


df2
```
